Answer Here

Dominance of EXW and FCA: The most prominent INCO Terms are EXW (Ex Works) and FCA (Free Carrier). This is a significant finding because these terms place the majority of the responsibility, including freight arrangement and risk, on the buyer (in this case, FedEx Logistics).

EXW: The seller makes the goods available at their own premises, and the buyer assumes all costs and risks from that point.

FCA: The seller delivers the goods to a carrier or another person nominated by the buyer at the seller's premises or another named place. The risk passes to the buyer at this point.

Limited Use of Other Terms: The chart shows that other terms like DAP (Delivered at Place), CIP (Carriage and Insurance Paid To), and DDP (Delivered Duty Paid) are used far less frequently. These terms would typically place more responsibility and risk on the vendor.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code

# Select numerical columns for correlation
numerical_cols = [
    'Line Item Quantity',
    'Line Item Value',
    'Pack Price',
    'Unit Price',
    'Weight (Kilograms)',
    'Freight Cost (USD)',
    'Line Item Insurance (USD)'
]
df_numerical = data[numerical_cols]

# Calculate the correlation matrix
correlation_matrix = df_numerical.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    mask=mask
)
plt.title('Correlation Heatmap of Key Numerical Variables')
plt.tight_layout()
# plt.savefig('/content/drive/MyDrive/FedEx Logistics/correlation_heatmap.png')
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer Here.

This heatmap shows how different factors, such as cost, weight, and quantity, are related to each other. The values on the chart represent the correlation coefficient, with values closer to 1 indicating a strong positive relationship, values closer to -1 indicating a strong negative relationship, and values around 0 indicating no relationship.

##### 2. What is/are the insight(s) found from the chart?

Answer Here

Key Insights from the Heatmap
Strong Positive Correlations:

Line Item Value & Line Item Insurance (0.88): This is a very strong relationship, showing that as the value of the goods increases, so does the cost of insurance.

Line Item Quantity & Line Item Value (0.81): This is a strong, expected correlation. As the number of items in a shipment increases, the total value of the shipment also increases.

Unit Price & Pack Price (0.98): This is a near-perfect correlation, which is logical since the pack price is directly derived from the unit price.

Moderate Positive Correlations:

Freight Cost (USD) & Weight (Kilograms) (0.64): This correlation indicates that as the weight of a shipment increases, the freight cost tends to increase as well.

Freight Cost (USD) & Line Item Value (0.50): There is a moderate relationship between freight cost and the value of a shipment. This suggests that more valuable shipments often incur higher shipping costs, but not always.

#### Chart - 15 - Pair Plot

````python
# Pair Plot visualization code

numerical_columns = [
    'Line Item Quantity',
    'Line Item Value',
    'Pack Price',
    'Unit Price',
    'Weight (Kilograms)',
    'Freight Cost (USD)',
    'Line Item Insurance (USD)'
]
numerical_df = data[numerical_columns]

# Drop rows with NaN values for a cleaner plot
numerical_df = numerical_df.dropna()

# Sample a subset of the data for better plot readability and performance
# Let's use 3000 rows as a good compromise.
if len(numerical_df) > 3000:
    sampled_df = numerical_df.sample(n=3000, random_state=42)
else:
    sampled_df = numerical_df

# Create the pair plot
sns.pairplot(sampled_df)

# Set the title and save the plot
plt.suptitle('Pair Plot of Key Numerical Variables', y=1.02, fontsize=16)
plt.tight_layout()
# plt.savefig('/content/drive/MyDrive/FedEx Logistics/pair_plot.png')
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer Here.

A pair plot is a grid of charts where each variable's distribution is shown on the diagonal, and the scatter plots for each combination of two variables are shown in the off-diagonal cells. This visualization helps to identify correlations, patterns, and outliers at a glance.

##### 2. What is/are the insight(s) found from the chart?

Answer Here

Key insights from the pair plot:

Variable Distributions: The histograms on the diagonal show that most variables, such as Freight Cost and Line Item Value, are highly right-skewed. This means that while most shipments have lower values and costs, a small number of very large and expensive shipments exist.

Relationships Between Variables: The scatter plots provide a visual confirmation of the relationships we previously saw in the correlation heatmap.

There's a clear strong positive linear relationship between Line Item Value and Line Item Quantity, and also between Line Item Insurance and Line Item Value, which is expected.

You can also see a positive trend between Freight Cost and Weight, showing that heavier shipments tend to have higher shipping costs.
